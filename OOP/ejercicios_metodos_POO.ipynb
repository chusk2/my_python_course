### Ejercicio 1 — Temperatura 🌡️
**Tipo:** Estático y de instancia

Crea una clase `Temperatura` que:
- Guarde un valor en **Celsius** como atributo de instancia.
- Tenga un **método estático** `celsius_a_fahrenheit(c)` que convierta y retorne el valor. `°F = (°C × 9/5) + 32`
- Tenga un **método de instancia** `mostrar()` que imprima el valor en Celsius y su equivalente en Fahrenheit (usando el método estático).

```python
# Uso esperado:
t = Temperatura(100)
t.mostrar()
# → 100°C equivale a 212.0°F
```

````python
class Temperature:
    def __init__(self, celcius):
        self.celcius = celcius
    
    @staticmethod
    def celcius_to_fahrenheit(celcius):
        return celcius * 9.5 + 32
    
    def show(self):
        return f"Temperatura: {self.celcius} ºC, {Temperature.celcius_to_fahrenheit(self.celcius):.2f} ºF"
````

````python
temp = Temperature(25)
temp.show()
````

### Ejercicio 2 — Contador de objetos 🔢
**Tipo:** De clase e instancia

Crea una clase `Robot` que:
- Lleve la cuenta de cuántos robots se han creado usando un **atributo de clase** `total`.
- Incremente ese contador en `__init__`.
- Tenga un **método de clase** `cuantos_robots()` que retorne el total actual.
- Tenga un **método de instancia** `presentarse()` que imprima el nombre del robot.

```python
# Uso esperado:
r1 = Robot("R2D2")
r2 = Robot("C3PO")
Robot.cuantos_robots()   # → 2
r1.presentarse()         # → "Soy R2D2"
```

````python
class Robot:
    robot_counter = 0
    def __init__(self, name):
        # incrementar el contador de robots creados
        Robot.robot_counter += 1
        # establecer nombre
        self.name = name
    
    @classmethod
    def robots_count(cls):
        return cls.robot_counter
    
    def introduce(self):
        print(f"Mi nombre es {self.name} y soy un robot.")
````

````python
r1 = Robot("R2D2")
r2 = Robot("C3PO")
Robot.robots_count()   # → 2
````

````python
r1.introduce()
````

### Ejercicio 3 — Constructor alternativo 📅
**Tipo:** De clase

Crea una clase `Date` con atributos `day`, `month` y `year`. Añade:
- Un **método de clase** `from_string(string)` que reciba una cadena con formato `"DD-MM-AAAA"` y retorne una instancia de `Date`.
- Un **método de instancia** `show()` que imprima la fecha con formato legible.

```python
# Uso esperado:
d = Date.from_string("25-12-2024")
d.show()   # → "25 de 12 de 2024"
```

````python
class Date:
    def __self__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    
    @classmethod
    def from_string(string):
        day, month, year = [int(i) for i in string.split("-")]
        return Date(day, month, year)
    
    def show(self):
        # preparar ordinal
        if str(self.day).endswith("1"):
            day = str(self.day) + "st"
        elif str(self.day).endswith("2"):
            day = str(self.day) + "nd"
        elif str(self.day).endswith("3"):
            day = str(self.day) + "rd"
        else:
            day = str(self.day) + "th"

        return f"{day} {self.month}, {self.year}"
````

````python
d = Date.from_string("25-12-2024")
d.show()   # → "25 de 12 de 2024"
````
