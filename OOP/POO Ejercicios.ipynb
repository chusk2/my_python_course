### Ejercicio 1.1 — La bebida

Crea una clase `Bebida` con los atributos `nombre`, `tamaño` (pequeño, mediano, grande) y `precio`. Instancia al menos tres bebidas distintas e imprime sus datos con un método `describir()`.

```python
class Bebida:
    def __init__(self, name, size, price):
        self.name = name
        self.size = size
        self.price = price
    
    def describir(self):
        return f"Bebida: {self.name}. Tamaño: {self.size}. Precio: {self.price} €"
        
```

```python
cafe = Bebida("Café con leche", "mediano", 2.50)
zumo = Bebida("Zumo de naranja", "grande", 3.00)
espresso = Bebida("Espresso", "pequeño", 1.20)

bebidas = [cafe, zumo, espresso]
```

```python
for bebida in bebidas:
    print(bebida.describir())
```

### Ejercicio 1.2 — La receta

Crea una clase `Receta` con los atributos `nombre`, `ingredientes` (una lista) y `tiempo_preparacion` (en minutos). Añade un método `es_rapida()` que devuelva `True` si se prepara en menos de 5 minutos. Crea varias recetas y filtra las rápidas.

```python
class Receta:
    def __init__(self, name: str, ingredients: list, tiempo_preparacion: int):
        self.name = name
        self.ingredients = ingredients
        self.tiempo_preparacion = tiempo_preparacion
    
    def es_rapida(self):
        # (explicar forma simplificada de devolver booleano)
        if self.tiempo_preparacion < 5:
            return True
        else:
            return False 
```

```python
recetas = {
    "pasta": Receta("Pasta boloñesa", ["pasta", "carne picada", "tomate", "cebolla", "ajo"], 35),

    "granola": Receta("Granola con yogur", ["yogur", "granola", "miel"], 3),
    
    "tortilla": Receta("Tortilla de patatas", ["huevos", "patatas", "cebolla", "aceite"], 25),
    
    "cafe_instantaneo": Receta("Café instantáneo", ["agua", "café soluble", "azúcar"], 2),

    "arroz": Receta("Arroz con leche", ["arroz", "leche", "azúcar", "canela"], 40),

    "lasaña": Receta("Lasaña de verduras", ["pasta", "berenjena", "tomate", "bechamel"], 60),

    "tostada": Receta("Tostada con aguacate", ["pan", "aguacate", "sal", "limón"], 4),

    "pollo": Receta("Pollo al horno", ["pollo", "patatas", "romero", "ajo", "aceite"], 50)
}
```

Lista las recetas rápidas:

```python
for recipe in recetas.values():
    if recipe.es_rapida():
        print(f"Receta: {recipe.name.lower()}. Tiempo de preparación: {recipe.tiempo_preparacion} minutos.")        
```

## 3. Métodos de instancia, de clase y estáticos

### Ejercicio 3.1 — La máquina de café

Crea una clase `MaquinaCafe` cuyo `__init__` reciba `marca`, `modelo` y `capacidad_ml` (capacidad del depósito en mililitros). Implementa:

- Un **método de instancia** `preparar(bebida, cantidad_ml)` que compruebe si hay suficiente capacidad en el depósito, descuente la cantidad usada e imprima qué bebida está preparando esa máquina.
- Un **método de clase** `crear_estandar()` que devuelva una instancia con valores por defecto (marca="Genérica", modelo="Básica", capacidad_ml=1000).
- Un **método estático** `convert_oz_to_ml(oz)` que convierta onzas a mililitros (1 oz ≈ 29.57 ml). Útil cuando una receta viene en onzas y necesitas saber cuántos ml descontar del depósito.

```python
class CoffeeMachine:
    def __init__(self, brand, model, capacity_ml):
        self.brand = brand
        self.model = model
        self.capacity_ml = capacity_ml
    
    # método de instancia
    def prepare(self, drink, volume_ml):

        # comprobar si hay volumen necesario para preparar la bebida
        if self.capacity_ml >= volume_ml:
            self.capacity_ml -= volume_ml
            print(f"Preparando bebida: {drink}...")
        else:
            print("La máquina no tiene suficiente volumen de agua para preparar la bebida solicitada. Por favor, rellene el depósito antes de volver a solicitar la bebida.")
    
    # método de clase
    def create_standard():
        return CoffeeMachine(brand="Generic", model="Basic", capacity_ml = 1000)

    # método estático
    def convert_oz_to_ml(oz):
        # factor de conversión: 1 oz ≈ 29.57 ml
        return oz / 29.57
```

```python
maquina = CoffeeMachine.create_standard()
oz_necesarias = 8
ml = CoffeeMachine.convert_oz_to_ml(oz_necesarias)  # ≈ 236.56 ml
maquina.prepare("Americano", ml)
```
