```python
def show_reservations(province, weekday):
    reservations_info = reservations[province]
    restaurant = reservations_info["nombre"]
    guests_count = reservations_info[weekday]
    return f"Nuestro restaurante situado en {province.title()} tiene confirmadas {guests_count} reservas para el {weekday.title()}."
```
