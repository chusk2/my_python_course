### Ejercicio 2: Filtrado por columna
**Tema:** CSV, condicionales, listas

Lee `transacciones_tienda.csv` y filtra solo las transacciones con `estado == 'completada'`. Imprime cuántas hay.

**Solución esperada:** Un número entre 30-40.

````python
with open("transacciones_tienda.csv", encoding="utf-8", mode="r") as file:
    reader = csv.DictReader(file)
    # complete transactions counter
    counter = 0
    for transaction in reader:
        if transaction["estado"] == "completada":
            print(transaction)
            counter += 1
print(f"Hay {counter} transacciones completadas.")
````

### Ejercicio 3: Suma de valores numéricos
**Tema:** CSV, conversión de tipos, bucles

Lee `transacciones_tienda.csv` y suma el total de dinero en la columna `monto`. Imprime el resultado.

**Solución esperada:** Un número aproximado a 5000-8000.

````python
with open("transacciones_tienda.csv", encoding="utf-8", mode="r") as file:
    reader = csv.DictReader(file)
    # total_amount
    total_amount = 0
    for transaction in reader:
        total_amount += float(transaction["monto"])

print(f"El total de las transacciones asciende a {total_amount:.2f} euros.")
````

### Ejercicio 5: Escritura de datos filtrados
**Tema:** CSV lectura y escritura, bucles

Lee `transacciones_tienda.csv`, filtra solo las transacciones del mes '03' (columna `mes`), y escribe el resultado en un nuevo archivo `transacciones_marzo.csv`.

**Solución esperada:** Un archivo CSV con ~8-12 transacciones de marzo.

````python
with open("transacciones_tienda.csv", encoding="utf-8", mode="r") as file:
    reader = csv.DictReader(file)
    march_transactions = []
    headers = reader.fieldnames
    # (explicar refactorización usando list comprehension)
    for transaction in reader:
        if int(transaction["mes"]) == 3:
            # guarda diccionarios
            march_transactions.append(transaction)
````

````python
march_transactions
````

````python
with open("transacciones_tienda_marzo.csv",
          encoding="utf-8", mode="w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=headers)
    writer.writeheader()
    writer.writerows(march_transactions)
````

### Ejercicio 8: Usuarios activos por rango de edad
**Tema:** CSV, condicionales compuestos, diccionarios

Lee `usuarios_registro.csv` y agrupa usuarios por rango de edad:
- 18-25
- 26-40
- 41+

Escribe el resultado en un archivo `usuarios_por_rango.csv` con columnas: `rango_edad, cantidad_usuarios`.

**Solución esperada:** Un CSV con 3 filas (los 3 rangos) y sus conteos.

````python
with open("usuarios_registro.csv", encoding="utf-8", mode="r") as file:
    reader = csv.DictReader(file)
    # diccionario para hacer conteo
    ages = {k:0 for k in ["18-25", "26-40", "41+"]}
    # lee la columna edad
    for line in reader:
        if 18 <= int(line["edad"]) <= 25:
            ages["18-25"] += 1
        elif 26 <= int(line["edad"]) <= 40:
            ages["26-40"] += 1
        else:
            ages["41+"] += 1

ages
````

````python
with open("usuarios_por_rango.csv", encoding = "utf-8", mode = "w") as file:
    writer = csv.DictWriter(file, fieldnames=ages.keys())
    writer.writeheader()
    writer.writerow(ages)
````

### Ejercicio 9: Detección de anomalías en logs
**Tema:** CSV, lógica compleja, filtrado múltiple

Lee `servidores_logs.csv` y detecta "anomalías" definidas como:
- Líneas con `nivel == 'ERROR'` Y `tiempo_respuesta > 5000` (ms)

Escribe las anomalías detectadas en un archivo `anomalias.csv` preservando todas las columnas originales.

**Solución esperada:** Un CSV con 3-8 anomalías detectadas.

````python
with open("servidores_logs.csv", encoding="utf-8", mode="r") as file:
    reader = csv.reader(file)
    with open("anomalias.csv", encoding="utf-8", mode="w") as f:
        writer = csv.writer(f)
        # escribe el header
        writer.writerow( next(reader) )
    # lee la columnas 3 (nivel) y 4 (tiempo_respuesta)
        for line in reader:
            # escribe los header
            level = line[2]
            response_time = line[3]
            if level == "ERROR" and int(response_time) > 5E3:
                writer.writerow(line)
````
